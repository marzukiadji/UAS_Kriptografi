```python
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Menghubungkan dengan Google Drive
drive.mount('/content/drive')

def calculate_image_measurements(image_path):
    # Langkah 1: Baca citra menggunakan OpenCV
    image = cv2.imread(image_path)

    # Pastikan citra berhasil dibaca
    if image is None:
        raise ValueError("Citra tidak dapat dibaca. Pastikan path citra benar.")

    # Langkah 2: Hitung dimensi citra (lebar, tinggi, dan jumlah saluran warna)
    height, width, channels = image.shape
    print(f"Dimensi Citra: Tinggi = {height}, Lebar = {width}, Saluran Warna = {channels}")

    # Langkah 3: Hitung luas citra
    area = width * height
    print(f"Luas Citra: {area} piksel")

    # Langkah 4: Hitung intensitas rata-rata per saluran warna
    average_intensity = np.mean(image, axis=(0, 1))
    print("Intensitas Rata-rata per Saluran Warna:", average_intensity)

    # Langkah 5: Hitung histogram per saluran warna
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
    print("Histogram Saluran Warna Biru:")
    for i, value in enumerate(histogram):
        print(f"Bin {i}: {int(value)} piksel")

    # Langkah 6: Tampilkan citra dan plot histogram
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Citra Asli")

    plt.subplot(2, 2, 2)
    plt.plot(average_intensity)
    plt.title("Intensitas Rata-rata per Saluran Warna")

    plt.subplot(2, 2, 3)
    plt.hist(image[:, :, 0].ravel(), bins=256, color='blue', alpha=0.7, rwidth=0.8)
    plt.title("Histogram Saluran Warna Biru")

    plt.show()

# Langkah 7: Panggil fungsi untuk mengukur citra
image_path = "/content/drive/MyDrive/codingan.jpeg"
calculate_image_measurements(image_path)
```
